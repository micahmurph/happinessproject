##World Happiness Report 

The World Happiness Report is a data set obtained from kaggle.com and reports data gathered from the Gallup World Poll from 2015-2019.  Factors like life-expectancy, economy, and freedom are all used to determine a happiness score and thus rank countries by that score.

The questions we wanted to answer were:

        What country will be the happiest in 2020?
        
        
        What factors impact happiness the most and least?
        
            For this question we used a horizontal bar graph from the matplotlib library to show the top and bottom countries and the corresponding factors that gave them their scores in proportion to their score.  This visualization highlights the largest and smallest contributor to the score.
            
            
        What continent had the happiest countries over the last five years?
        
            Using pandas, we grouped the countries and averaged their happiness scores over the five years to obtain the highest averages.

```python
file_name = '2015.csv'
data_2015 = pd.read_csv(file_name)
##data_2015.head(1)

data_2015.drop(['Standard Error','Region', 'Dystopia Residual'], axis=1, inplace=True)

##df = pd.DataFrame({"A": [1, 2, 3], "B": [4, 5, 6]})


data_2015.rename(columns={"Happiness Rank": "Overall rank", "Happiness Score": "Overall score", "Economy (GDP per Capita)": "GDP per capita", "Family": "Social support", "Health (Life Expectancy)":"Healthy life expectancy", "Freedom":"Freedom to make life choices", "Trust (Government Corruption)":"Perceptions of corruption"}, inplace=True)
data_2015['Year'] = '2015'
data_2015.head(1)
```

```python
data_2016 = pd.read_csv('2016.csv')
data_2016.drop(['Region', 'Lower Confidence Interval', 'Upper Confidence Interval', 'Dystopia Residual'], axis=1, inplace=True)
data_2016.rename(columns={"Happiness Rank": "Overall rank", "Happiness Score": "Overall score", "Economy (GDP per Capita)": "GDP per capita", "Family": "Social support", "Health (Life Expectancy)":"Healthy life expectancy", "Freedom":"Freedom to make life choices", "Trust (Government Corruption)":"Perceptions of corruption"}, inplace=True)
data_2016['Year'] = '2016'
data_2016.head(1)
```

```python
data_2017 = pd.read_csv('2017.csv')
data_2017.drop(['Whisker.high', 'Whisker.low', 'Dystopia.Residual'], axis=1, inplace=True)
data_2017['Year'] = '2017'
data_2017.rename(columns={"Happiness.Rank": "Overall rank", "Happiness.Score": "Overall score", "Economy..GDP.per.Capita.": "GDP per capita", "Family": "Social support", "Health..Life.Expectancy.":"Healthy life expectancy", "Freedom":"Freedom to make life choices", "Trust..Government.Corruption.":"Perceptions of corruption"}, inplace=True)
data_2017.head(1)
```

```python
data_2018 = pd.read_csv('2018.csv')
data_2018.rename(columns={"Country or region": "Country", "Score":"Overall score"}, inplace=True)
data_2018['Year'] = '2018'
data_2018.head(1)
```

```python
data_2019 = pd.read_csv('2019.csv')
data_2019.rename(columns={"Country or region": "Country", "Score":"Overall score"}, inplace=True)
data_2019['Year'] = '2019'
data_2019.head(1)
```

```python
frames = [data_2015, data_2016, data_2017, data_2018, data_2019]

df = pd.concat(frames, sort=False)
##df.drop(['Overall Rank', 'Overall Score', 'GDP per Capita'], axis=1, inplace=True)
df.head(100)
```

```python
df.set_index('Country', inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.to_csv('combined.csv')
```

```python
#What factors impact happiness the most in the top ten countries?

import numpy as np
from matplotlib import pyplot as plt

df[['GDP per capita', 'Social support','Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']].head(10).plot(kind='barh',
xticks=np.arange(9), stacked=True, figsize= (10, 10))

plt.xlabel("Happiness Score")
plt.title('Happiness Score of the top 10 Countries')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
##What factors impact happiness the most in the bottom ten?

df[['GDP per capita', 'Social support','Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']].tail(10).plot(kind='barh',
xticks=np.arange(9), stacked=True, figsize= (10, 10))

plt.xlabel("Happiness Score")
plt.title('Happiness Score of the bottom 10 Countries')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
##What continent had the happiest countries in the last five years?


df.groupby('Country').mean()['Overall score'].sort_values().tail(10) 
```

Six out of ten of the countries with the highest happiness scores are European.
