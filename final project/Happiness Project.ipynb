```python
file_name = '2015.csv'
data_2015 = pd.read_csv(file_name)
##data_2015.head(1)

data_2015.drop(['Standard Error','Region', 'Dystopia Residual'], axis=1, inplace=True)

##df = pd.DataFrame({"A": [1, 2, 3], "B": [4, 5, 6]})


data_2015.rename(columns={"Happiness Rank": "Overall rank", "Happiness Score": "Overall score", "Economy (GDP per Capita)": "GDP per capita", "Family": "Social support", "Health (Life Expectancy)":"Healthy life expectancy", "Freedom":"Freedom to make life choices", "Trust (Government Corruption)":"Perceptions of corruption"}, inplace=True)

data_2015.head(1)
```

```python
data_2016 = pd.read_csv('2016.csv')
data_2016.drop(['Region', 'Lower Confidence Interval', 'Upper Confidence Interval', 'Dystopia Residual'], axis=1, inplace=True)
data_2016.rename(columns={"Happiness Rank": "Overall rank", "Happiness Score": "Overall score", "Economy (GDP per Capita)": "GDP per capita", "Family": "Social support", "Health (Life Expectancy)":"Healthy life expectancy", "Freedom":"Freedom to make life choices", "Trust (Government Corruption)":"Perceptions of corruption"}, inplace=True)

data_2016.head(1)
```

```python
data_2017 = pd.read_csv('2017.csv')
data_2017.drop(['Whisker.high', 'Whisker.low', 'Dystopia.Residual'], axis=1, inplace=True)

data_2017.rename(columns={"Happiness.Rank": "Overall rank", "Happiness.Score": "Overall score", "Economy..GDP.per.Capita.": "GDP per capita", "Family": "Social support", "Health..Life.Expectancy.":"Healthy life expectancy", "Freedom":"Freedom to make life choices", "Trust..Government.Corruption.":"Perceptions of corruption"}, inplace=True)
data_2017.head(1)
```

```python
data_2018 = pd.read_csv('2018.csv')
data_2018.rename(columns={"Country or region": "Country", "Score":"Overall score"}, inplace=True)
data_2018.head(1)
```

```python
data_2019 = pd.read_csv('2019.csv')
data_2019.rename(columns={"Country or region": "Country", "Score":"Overall score"}, inplace=True)
data_2019.head(1)
```

```python
frames = [data_2015, data_2016, data_2017, data_2018, data_2019]

df = pd.concat(frames, sort=False)
df.drop(['Overall Rank', 'Overall Score', 'GDP per Capita'], axis=1, inplace=True)
df.head(100)
```
